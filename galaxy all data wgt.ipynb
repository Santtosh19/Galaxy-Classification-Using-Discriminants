```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,confusion_matrix

#This is where we put our csv file from the directory of 'bnnz/output/test_singleCls_quick/clasif/eval/ANNZ_singleCls_0000.csv'
#In this case i'm using jupyter notebook, so i will import the csv file into the same file where im doing this notebook
# Load your CSV file
file_path = 'annh2_1000.csv'
data = pd.read_csv(file_path)

predicted = data['F :ANNZ_0']
#ACTUALLY I PUT 1:NO BULGE AND 0: ROUNDED
#TO AVOID CONFUSION 
data['type'] = data['type'].replace({1: 1, 0: 0})
true_labels = data['type']
threshold = 0.5
binary_predictions = (predicted >= threshold).astype(int)

accuracy = accuracy_score(true_labels, binary_predictions)
precision = precision_score(true_labels, binary_predictions)
recall = recall_score(true_labels, binary_predictions)
f1_score = f1_score(true_labels, binary_predictions)
confusion_matrix=confusion_matrix(true_labels,binary_predictions)

print("Model Metrics:")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1_score}")
print(f"Confusion_Matrix:\n{confusion_matrix}")

#Statistics Calculation
from sklearn.metrics import confusion_matrix, mean_absolute_error
import numpy as np
tn, fp, fn, tp = confusion_matrix(true_labels,binary_predictions).ravel() #ravel() is something like .items() but this one is for dictionary : key,value
# ravel() makes it into 1d array and we can unpack
errors = [fp, fn]
positive=[tn,tp]
# Standard deviation 
std_dev = np.std(errors)
std_dev_p=np.std(positive)
# Mean absolute error
mae = mean_absolute_error(true_labels, binary_predictions)

print (f"tn:{tn}")
print(f"fp:{fp}")
print(f"fn:{fn}")
print(f"tp:{tp}")
print(f"Standard Deviation of Errors: {std_dev}")
print(f"Standard Deviation of Positives: {std_dev_p}")
print(f"Mean Absolute Error: {mae}")


#####################VISUALIZATION
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.figure(figsize=(8,8))
sns.kdeplot(data[true_labels == 1]['F :ANNZ_0'], color='blue', label='No Bulge Galaxy (Class 1)', lw=3)##lw is line width
sns.kdeplot(data[true_labels == 0]['F :ANNZ_0'], color='red', label='Rounded Galaxy (Class 0)', lw=3)
plt.axvline(x=0.5, color='green', linestyle='--', label='Decision Threshold (0.5)')
plt.title(f'Distribution of Predictions using Gaussian KDE')
plt.xlabel('Predictions')
plt.ylabel('Density')
plt.legend()
plt.show()


#########OUTLINES
plt.figure(figsize=(8, 8))
plt.scatter(data.index[true_labels == 1], predicted[true_labels == 1], c='blue', alpha=0.5, s=25, linewidth=0.5, label='No Bulge Galaxy : 1')
plt.scatter(data.index[true_labels == 0], predicted[true_labels == 0], c='red', alpha=0.5, s=25, linewidth=0.5, label='Rounded Galaxy : 0')
plt.axhline(y=threshold, color='green', linestyle='--', label='Decision Threshold (0.5)')
plt.title('Scatter Plot of Predicted Data Points')
plt.xlabel('Data Points Index')
plt.ylabel('Predictions')
plt.legend()
plt.show()
```
